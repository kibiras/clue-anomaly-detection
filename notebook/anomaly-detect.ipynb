````python
from sklearn import metrics

def classification_metrics(y_true, y_pred):
    """
    Calculate and print various classification metrics.
    
    Args:
    - y_true (Series): True labels.
    - y_pred (Series): Predicted labels.
    
    Returns:
    - None (prints the metrics)
    """
    
    y_true = y_true.astype('int64')
    
    # Calculate the confusion matrix
    tn, fp, fn, tp = metrics.confusion_matrix(y_true, y_pred).ravel()

    # Calculate the metrics
    tpr = tp / (tp + fn)
    fpr = fp / (fp + tn)
    tnr = tn / (tn + fp)
    prec = tp / (tp + fp)
    rec = tpr
    f1 = 2 * (prec * rec) / (prec + rec)
    acc = (tp + tn) / (tp + tn + fp + fn)  

    results = {
        'TP': tp,
        'FP': fp,
        'TN': tn,
        'FN': fn,
        'TPR/Rec': tpr,
        'FPR': fpr,
        'TNR': tnr,
        'Prec': prec,
        'F1': f1,
        'ACC': acc,
    }

    return results
````

# Baseline

````python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize
import numpy as np

# Initialize a dictionary to hold the user profiles and another one to hold the similarity scores
user_profiles = {}
similarity_scores = {}

# Initialize a list to hold the anomaly labels
anomalies = []

# Smoothing factor for profile updates
alpha = 0.1

# Iterate over the rows of df_a and df_bq
for (_, row_a), (_, row_b) in zip(df_a.iterrows(), df_e.iterrows()):
    uid = row_a['uid']
    event_embeddings = row_b.values.reshape(1, -1)
    event_embeddings = normalize(event_embeddings)  # ensure the embeddings are normalized

    # If the user profile doesn't exist, create it
    if uid not in user_profiles:
        user_profiles[uid] = event_embeddings
        similarity_scores[uid] = [1]  # The initial similarity is 1
        anomalies.append(0)
    else:
        # Calculate the cosine similarity between the user profile and the event embeddings
        similarity = cosine_similarity(user_profiles[uid], event_embeddings)[0,0]
        similarity_scores[uid].append(similarity)

        # Calculate dynamic threshold (here, lower quartile of the historical similarity scores)
        dynamic_threshold = np.percentile(similarity_scores[uid], 25)

        # If the cosine similarity is less than the dynamic threshold, label the event as an anomaly
        if similarity < dynamic_threshold:
            anomalies.append(1)
        else:
            anomalies.append(0)

        # Update the user profile with exponential decay
        user_profiles[uid] = alpha * event_embeddings + (1 - alpha) * user_profiles[uid]

# Add the anomaly labels to df_a
df_a['anomaly_predicted'] = anomalies
````

````python
classification_metrics(df_a['anomaly'], df_a['anomaly_predicted'])
````
